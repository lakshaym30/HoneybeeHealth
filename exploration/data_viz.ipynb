```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

inspections = pd.read_csv("../data/derived/Inspections_With_Weather.csv")
sns.countplot(x='Healthy', data=inspections)
plt.title('Class Distribution: Hive Health')
plt.show()
```

```python
sns.histplot(inspections['Temperature'], kde=True)
plt.title('Temperature Distribution')
plt.show()
```

```python
sns.boxplot(x='Healthy', y='Temperature', data=inspections)
plt.title('Temperature by Hive Health')
plt.show()
```

```python
sns.violinplot(x='Healthy', y='Humidity', data=inspections)
plt.title('Humidity by Hive Health')
plt.show()
```

```python
numeric_df = inspections.select_dtypes(include='number').drop(['InspectionID', 'HiveID', 'Percent_Met', 'ApiaryID', 'StationID'], axis=1)
corr = numeric_df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```
